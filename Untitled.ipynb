Pie chart for the Total Students DropOut Rate from the Secondary School

```python
# prompt: Using dataframe div_2: pie with values with title

import altair as alt
chart = alt.Chart(div_2).mark_arc().encode(
    theta=alt.Theta(field="Total", type="quantitative"),
    color=alt.Color(field="India/State/UT", type="nominal"),
    tooltip=['India/State/UT', 'Total']
).properties(
    title='Total Students Dropout Rate from Secondary(9-10) Education'
)
chart
```
